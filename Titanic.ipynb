```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

data_train = pd.read_csv("C:\\Users\\Bosco\\Favorites\\train.csv")
data_train
```

```python
data_train.info()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
fig.set(alpha=0.2)  # è®¾å®šå›¾è¡¨é¢œè‰²alphaå‚æ•°

plt.subplot2grid((2,3),(0,0))             # åœ¨ä¸€å¼ å¤§å›¾é‡Œåˆ†åˆ—å‡ ä¸ªå°å›¾
data_train.Survived.value_counts().plot(kind='bar')# æŸ±çŠ¶å›¾ 
plt.title(u"Survived Status (1)") # æ ‡é¢˜
plt.ylabel(u"nums")  

plt.subplot2grid((2,3),(0,1))
data_train.Pclass.value_counts().plot(kind="bar")
plt.ylabel(u"nums")
plt.title(u"passenger level")

plt.subplot2grid((2,3),(0,2))
plt.scatter(data_train.Survived, data_train.Age)
plt.ylabel(u"age")                         # è®¾å®šçºµåæ ‡åç§°
plt.grid(b=True, which='major', axis='y') 
plt.title(u"Survived by age (1)")


plt.subplot2grid((2,3),(1,0), colspan=2)
data_train.Age[data_train.Pclass == 1].plot(kind='kde')   
data_train.Age[data_train.Pclass == 2].plot(kind='kde')
data_train.Age[data_train.Pclass == 3].plot(kind='kde')
plt.xlabel(u"age")# plots an axis lable
plt.ylabel(u"density") 
plt.title(u"All level passenger with age")
plt.legend((u'First class', u'Second class',u'Third class'),loc='best') # sets our legend for our graph.


plt.subplot2grid((2,3),(1,2))
data_train.Embarked.value_counts().plot(kind='bar')
plt.title(u"Number of board")
plt.ylabel(u"nums")  
plt.show()
```

```python
# å„ä¹˜å®¢ç­‰çº§çš„èŽ·æ•‘æƒ…å†µ
fig = plt.figure()
fig.set(alpha=0.2)  # è®¾å®šå›¾è¡¨é¢œè‰²alphaå‚æ•°

Survived_0 = data_train.Pclass[data_train.Survived == 0].value_counts()
Survived_1 = data_train.Pclass[data_train.Survived == 1].value_counts()
df = pd.DataFrame({u'Survived':Survived_1, u'Unsurvived':Survived_0})
df.plot(kind='bar', stacked=True)
plt.title(u"Survived by level")
plt.xlabel(u"Passenger Level") 
plt.ylabel(u"nums") 
plt.show()
```

```python
# å„æ€§åˆ«çš„èŽ·æ•‘æƒ…å†µ
fig = plt.figure()
fig.set(alpha=0.2)  # è®¾å®šå›¾è¡¨é¢œè‰²alphaå‚æ•°

Survived_m = data_train.Survived[data_train.Sex == 'male'].value_counts()
Survived_f = data_train.Survived[data_train.Sex == 'female'].value_counts()
df=pd.DataFrame({u'male':Survived_m, u'female':Survived_f})
df.plot(kind='bar', stacked=True)
plt.title(u"Survived by sex")
plt.xlabel(u"sex") 
plt.ylabel(u"nums")
plt.show()
```

```python
# å„ç§èˆ±çº§åˆ«æƒ…å†µä¸‹å„æ€§åˆ«çš„èŽ·æ•‘æƒ…å†µ
fig = plt.figure()
fig.set(alpha=0.65) # è®¾ç½®å›¾åƒé€æ˜Žåº¦ï¼Œæ— æ‰€è°“
plt.title(u"Survived with level and sex")

ax1=fig.add_subplot(141)
data_train.Survived[data_train.Sex == 'female'][data_train.Pclass != 3].value_counts().plot(kind='bar', label="female, high class", color='#FA2479')
ax1.set_xticklabels([u"yes", u"no"], rotation=0)
ax1.legend([u"female/high level"], loc='best')

ax2=fig.add_subplot(142, sharey=ax1)
data_train.Survived[data_train.Sex == 'female'][data_train.Pclass == 3].value_counts().plot(kind='bar', label='female, low class', color='pink')
ax2.set_xticklabels([u"yes", u"no"], rotation=0)
plt.legend([u"female/low level"], loc='best')

ax3=fig.add_subplot(143, sharey=ax1)
data_train.Survived[data_train.Sex == 'male'][data_train.Pclass != 3].value_counts().plot(kind='bar', label='male, high class',color='lightblue')
ax3.set_xticklabels([u"no", u"yes"], rotation=0)
plt.legend([u"male/high level"], loc='best')

ax4=fig.add_subplot(144, sharey=ax1)
data_train.Survived[data_train.Sex == 'male'][data_train.Pclass == 3].value_counts().plot(kind='bar', label='male low class', color='steelblue')
ax4.set_xticklabels([u"no", u"yes"], rotation=0)
plt.legend([u"male/low level"], loc='best')

plt.show()
```

```python
# å„ç™»èˆ¹æ¸¯å£çš„èŽ·æ•‘æƒ…å†µ
fig = plt.figure()
fig.set(alpha=0.2)  # è®¾å®šå›¾è¡¨é¢œè‰²alphaå‚æ•°

Survived_0 = data_train.Embarked[data_train.Survived == 0].value_counts()
Survived_1 = data_train.Embarked[data_train.Survived == 1].value_counts()
df=pd.DataFrame({u'yes':Survived_1, u'no':Survived_0})
df.plot(kind='bar', stacked=True)
plt.title(u"Survived with Embarked")
plt.xlabel(u"Embarked") 
plt.ylabel(u"nums") 

plt.show()
```
